```python
import numpy as np
from keras.utils import to_categorical
```

```python
from keras.datasets import mnist
(x_train,y_train),(x_test,y_test)=mnist.load_data()
```

```python
x_train=np.array(x_train).astype('float32')/255
y_train=np.array(y_train)
x_test=np.array(x_test).astype('float32')/255
y_test=np.array(y_test)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
y_train=to_categorical(y_train)
y_test=to_categorical(y_test)
```

```python
from keras import models,layers
from keras.layers import Dense, Conv2D, Flatten
model =models.Sequential([
    Conv2D(filters=96, kernel_size=(3,3), strides=(2,2), activation='relu', input_shape=(28,28,1)),
    layers.BatchNormalization(),
    layers.MaxPool2D(pool_size=(2,2), strides=(2,2)),
    Conv2D(filters=256, kernel_size=(3,3), strides=(1,1), activation='relu', padding="same"),
    layers.BatchNormalization(),
    layers.MaxPool2D(pool_size=(2,2), strides=(2,2)),
    Conv2D(filters=384, kernel_size=(3,3), strides=(1,1), activation='relu', padding="same"),
    layers.BatchNormalization(),
    Conv2D(filters=384, kernel_size=(3,3), strides=(1,1), activation='relu', padding="same"),
    layers.BatchNormalization(),
    Conv2D(filters=256, kernel_size=(3,3), strides=(1,1), activation='relu', padding="same"),
    layers.BatchNormalization(),
    layers.MaxPool2D(pool_size=(2,2), strides=(2,2)),
    Flatten(),
    Dense(1024, activation='relu'),
    layers.Dropout(0.5),
    Dense(1024, activation='relu'),
    layers.Dropout(0.5),
    Dense(10, activation='softmax')
])
```

```python
model.compile(optimizer="sgd",loss="categorical_crossentropy",metrics=['accuracy'])
model.summary()
```

```python
model.fit(x=x_train,y=y_train,batch_size=128,epochs=10)
```

```python
model.evaluate(x_test,y_test)
pred = model.predict(x_test)
```

```python
from sklearn.metrics import classification_report
labels = [np.argmax(i) for i in pred]
t_labels = [np.argmax(i) for i in y_test]
print(classification_report(t_labels, labels))
```
